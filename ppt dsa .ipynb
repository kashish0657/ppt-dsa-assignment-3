## 9:- A study is conducted to compare the effectiveness of two different medications . The recovery times ofthe patients in each group are as follows: Medication A:[10,12,14,11,13] Medication B:[15,17,16,14,18] Perform an analysis of variance to determine if there is a significant differences in mean recovery times between the two medications .

```python
ans:-

Step 1: State the null hypothesis (H0) and alternative hypothesis (H1):
H0: The mean recovery times of the two medications are equal.
H1: The mean recovery times of the two medications are not equal.

Step 2: Calculate the means and variances of each group:
Medication A: [10, 12, 14, 11, 13]
Mean A = (10 + 12 + 14 + 11 + 13) / 5 = 12
Variance A = ((10 - 12)^2 + (12 - 12)^2 + (14 - 12)^2 + (11 - 12)^2 + (13 - 12)^2) / 4 = 2.5

Medication B: [15, 17, 16, 14, 18]
Mean B = (15 + 17 + 16 + 14 + 18) / 5 = 16
Variance B = ((15 - 16)^2 + (17 - 16)^2 + (16 - 16)^2 + (14 - 16)^2 + (18 - 16)^2) / 4 = 2.5

Step 3: Calculate the overall mean and variance:
Overall mean = (Mean A + Mean B) / 2 = (12 + 16) / 2 = 14
Overall variance = ((10 - 14)^2 + (12 - 14)^2 + (14 - 14)^2 + (11 - 14)^2 + (13 - 14)^2 +
                    (15 - 14)^2 + (17 - 14)^2 + (16 - 14)^2 + (14 - 14)^2 + (18 - 14)^2) / 8 = 4.5

Step 4: Calculate the between-group sum of squares (SSbetween):
SSbetween = (5 * (Mean A - Overall mean)^2) + (5 * (Mean B - Overall mean)^2)
SSbetween = (5 * (12 - 14)^2) + (5 * (16 - 14)^2) = 40

Step 5: Calculate the within-group sum of squares (SSwithin):
SSwithin = (4 * Variance A) + (4 * Variance B)
SSwithin = (4 * 2.5) + (4 * 2.5) = 20

Step 6: Calculate the degrees of freedom:
Degrees of freedom between (dfbetween) = number of groups - 1 = 2 - 1 = 1
Degrees of freedom within (dfwithin) = total number of observations - number of groups = 10 - 2 = 8

Step 7: Calculate the F-statistic:
F-statistic = (SSbetween / dfbetween) / (SSwithin / dfwithin)
F-statistic = (40 / 1) / (20 / 8) = 16

Step 8: Determine the critical value of F:
We need to specify a significance level (α) to determine the critical value
of F. Let's assume α = 0.05.

Using an F-distribution table or statistical software, we can find the critical 
value of F with dfbetween = 1 and dfwithin = 8 at α = 0.05. The critical value of F is approximately 5.32.

Step 9: Compare the F-statistic with the critical value of F:
If the calculated F-statistic is greater than the critical value of F, we reject 
the null hypothesis. Otherwise, we fail to reject the null hypothesis.
```

The data collected is as follows: Group A:[4,3,5,2,4] Group B:[3,2,4,3,3] Perform a Mann-Whitney U test to detrmine if there is a significant difference in the median preferences between the two groups .

```python
ans:-

Step 1: State the null hypothesis (H0) and the alternative hypothesis (H1):
H0: There is no significant difference in the median preferences between
Group A and Group B.
H1: There is a significant difference in the median preferences between
Group A and Group B.

Step 2: Rank the data:
Combine the data from both groups and rank them in ascending order, 
assigning ranks based on their values.

Group A: [4, 3, 5, 2, 4] → Ranks: [3, 2, 5, 1, 3]
Group B: [3, 2, 4, 3, 3] → Ranks: [2, 1, 4, 2, 2]

Step 3: Calculate the Mann-Whitney U statistic:
U is calculated as the sum of ranks in the smaller sample (U1) minus the 
expected value of U1. The expected value of U1 is given by the formula:
E(U1) = (n1 * (n1 + n2 + 1)) / 2

Group A (n1 = 5) has U1 = 3 + 2 + 5 + 1 + 3 = 14
Group B (n2 = 5) has E(U1) = (5 * (5 + 5 + 1)) / 2 = 15

U = U1 - E(U1) = 14 - 15 = -1

Step 4: Calculate the critical value:
The critical value of U depends on the sample size. For a two-tailed
test at a significance level of α = 0.05, you can consult a Mann-Whitney
U critical values table or use a statistical software.

Given that the total sample size (n) is 10, the critical value for U at α = 0.05 is 12.

Step 5: Compare the absolute value of U with the critical value:
|U| = |-1| = 1 < 12

Step 6: Make a decision:
Since the absolute value of U is less than the critical value, we fail to 
reject the null hypothesis. There is insufficient evidence to conclude that 
there is a significant difference in the median preferences between Group A and Group B.
```

```python
ans:-

Step 1: State the hypotheses:

Null hypothesis (H0): There is no significant difference in the median
accuracy scores between the algorithms.
Alternative hypothesis (Ha): There is a significant difference in the 
median accuracy scores between the algorithms.
Step 2: Set the significance level (α):
Choose a significance level (α) to determine the threshold for statistical 
significance. Let's assume α = 0.05, which is a common choice.

Step 3: Rank the data:
Combine all the accuracy scores from the three algorithms into a single list
and assign ranks to each value. Ties are handled by assigning the average rank.

Algorithm A: [0.85, 0.80, 0.82, 0.87, 0.83]
Algorithm B: [0.78, 0.82, 0.84, 0.80, 0.79]
Algorithm C: [0.90, 0.88, 0.89, 0.86, 0.87]

Combined ranked data: [0.78, 0.80, 0.82, 0.83, 0.84, 0.86, 0.87, 0.88, 0.89, 0.90]
Ranks: [1, 2.5, 4, 5, 6, 7, 8, 9, 10, 11]

Step 4: Calculate the test statistic:
The Kruskal-Wallis test statistic is calculated using the following formula:

H = (12 / (N * (N + 1))) * Σ(Rj - R̄)² / nj

Where:

N is the total number of observations (in this case, 10).
Rj is the sum of the ranks for the jth group.
R̄ is the mean of all the ranks.
nj is the number of observations in the jth group.
Calculate H for the given data:

H = (12 / (10 * (10 + 1))) * ((8 - 5.5)² / 5 + (25 - 5.5)² / 5 + (47 - 5.5)² / 5)
H = (12 / 110) * (8.5²/5 + 19.5²/5 + 41.5²/5)
H ≈ 2.23

Step 5: Determine the critical value:
To compare the test statistic with the critical value, you need to use the 
chi-square distribution with k - 1 degrees of freedom, where k is the number
of groups (in this case, 3).

Using a significance level (α) of 0.05, and k - 1 = 2 degrees of freedom, the
critical value is approximately 5.991.

Step 6: Compare the test statistic with the critical value:
If the test statistic (H) is greater than the critical value, we reject the null
hypothesis; otherwise, we fail to reject the null hypothesis.

In this case, H ≈ 2.23, which is less than the critical value of 5.991.

Step 7: Make a decision:
Since the test statistic (H) is less than the critical value, we fail to reject the
null hypothesis. There is not enough evidence to conclude that there is a significant
difference in the median accuracy scores between the algorithms.
```

## 20:- A company wants to analyze the relationship between advertising expenditure and sales.The data collected is as follows: Advertising expenditure :[10,15,12,8,14] Sales :[25,30,28,20,26] perform a multiple regression analysis to predict sales based on advertising expenditure . 

```python
import pandas as pd
import statsmodels.api as sm

# Create a pandas DataFrame with the advertising expenditure and sales data
data = {
    'Advertising Expenditure': [10, 15, 12, 8, 14],
    'Sales': [25, 30, 28, 20, 26]
}
df = pd.DataFrame(data)

# Add a constant term to the DataFrame
df = sm.add_constant(df)

# Create the regression model
model = sm.OLS(df['Sales'], df[['const', 'Advertising Expenditure']])

# Fit the model
results = model.fit()

# Print the summary of the regression analysis
print(results.summary())
```
